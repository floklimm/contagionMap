## Plot embeddings of noisy geometric ring lattice networks

```python
%load_ext autoreload
%autoreload 2

import cmap as conmap
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set_theme()
import pandas as pd

# For pretty colourmaps
import palettable
from matplotlib.colors import ListedColormap

sns.set_style("white")

from sklearn.decomposition import PCA
```

We construct the noisy ring lattice networks. The first network has only geometric edges and the second network has geometric and non-geometric edges.

```python
noisyRLSmall1 = conmap.constructNoisyRingLattice(numberNodes=40,geometricDegree=2,nongeometricDegree=0)
noisyRLSmall2 = conmap.constructNoisyRingLattice(numberNodes=40,geometricDegree=2,nongeometricDegree=2)
```

```python
fig = plt.figure(figsize=(16,8))


# A) plot manifold
ax = fig.add_subplot(131)
nodeAngle = np.arange(0,2*np.pi,2*np.pi/100)
nodeAngle=np.append(nodeAngle,0)
xCircle = np.sin(nodeAngle)
yCircle = np.cos(nodeAngle)
plt.plot(xCircle,yCircle,color='#808080',linewidth=5)
ax.set_xlim(-1.1,1.1)
ax.set_ylim(-1.1,1.1)
plt.title('manifold $\mathcal{M}$',size=30)


ax.set_aspect('equal', adjustable='box')
plt.axis('off')

# B) Plot ring lattice network (without non-geometric edges)
ax = fig.add_subplot(132)
conmap.drawGeometricNetwork(noisyRLSmall1,node_size=50)
plt.title('geometric lattice',size=30)

ax.set_aspect('equal', adjustable='box')
plt.axis('off')
ax.set_xlim(-1.1,1.1)
ax.set_ylim(-1.1,1.1)

# C) Plot noisy ring lattice network (with non-geometric edges)
ax = fig.add_subplot(133)
conmap.drawGeometricNetwork(noisyRLSmall2,node_size=50)
plt.title('noisy geometric lattice',size=30)
ax.set_aspect('equal', adjustable='box')
plt.axis('off')
ax.set_xlim(-1.1,1.1)
ax.set_ylim(-1.1,1.1)
plt.tight_layout()
plt.savefig('./figures/Fig2-exampledNoisyGeometricLattice.pdf')
```
